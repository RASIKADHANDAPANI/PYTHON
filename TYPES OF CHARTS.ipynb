# LINE CHARTS

### Simple line plots

```python
import matplotlib.pyplot as plt
import numpy as np
  
# define data values
x = np.array([1, 2, 3, 4])  # X-axis points
y = x*2  # Y-axis points
  
plt.plot(x, y)  # Plot the chart
plt.show()  # display
```

```python
plt.xlabel("X-axis")  # add X-axis label
plt.ylabel("Y-axis")  # add Y-axis label
plt.title("CHART 1 SHOWING X VALUE")  # add title
plt.plot(x, y,"-.")  # Plot the chart
plt.show()  # display
```

### Multiple charts (Two Different Graphs)

```python
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("CHART 1 SHOWING X VALUE")  
plt.plot(x, y)  # Plot the chart
plt.show()  # display

# The figure() function helps in creating a new figure that can hold a new chart in it.
plt.figure()

x1 = np.array([10, 20, 30, 40]) 
y1 = x*2
plt.xlabel("X1-axis")
plt.ylabel("Y1-axis")
plt.title("CHART 2 SHOWING X2 VALUE")  
plt.plot(x1, y1, '-.')
  
# Show another chart with '-' dotted line
plt.show()
```

```python
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("CHART 1 SHOWING X VALUE")  
plt.subplot(1,2,1)
plt.plot(x, y)  # Plot the chart

x1 = np.array([10, 20, 30, 40]) 
y1 = x*2
plt.xlabel("X1-axis")
plt.ylabel("Y1-axis")
plt.title("CHART 2 SHOWING X2 VALUE")  
plt.subplot(1,2,2)
plt.plot(x1, y1, '-.')
  
# Show another chart with '-' dotted line
plt.show()
```

### Multiple charts (In Same Graphs)

```python
x = np.array([1, 2, 3, 4])
y = x*2
  
# first plot with X and Y data
plt.plot(x, y)
  
x1 = np.array([10, 20, 30, 40]) 
y1 = x*2

# second plot with x1 and y1 data
plt.plot(x1, y1, '-.')
  
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('X AND X1 VALUES PLOTTED IN SAME GRAPH')
plt.show()
```

```python
year = [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
unemployment_rate = [9.8, 12, 8, 7.2, 6.9, 7, 6.5, 6.2, 5.5, 6.3]
  
plt.plot(year, unemployment_rate)
plt.title('unemployment rate vs year')
plt.xlabel('year')
plt.ylabel('unemployment rate')
plt.show()
```

### maker styles






```python
#By Applying styles

plt.plot(year, unemployment_rate, color='red', marker='_')
plt.title('unemployment rate vs year', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('unemployment rate', fontsize=14)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
data = {'year': [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010],
        'unemployment_rate': [9.8, 12, 8, 7.2, 6.9, 7, 6.5, 6.2, 5.5, 6.3]
       }
  
df = pd.DataFrame(data)
plt.plot(df['year'], df['unemployment_rate'], color='red', marker='o')
plt.title('unemployment rate vs year', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('unemployment rate', fontsize=14)
plt.grid(True)
plt.show()
```

# HISTOGRAM

This method uses **numpy.histogram** to bin the data in x and count the number of values in each bin, then draws the distribution e

```python
a = np.random.randint(9,size=20)
print(a)
plt.hist(a)
plt.show()
```

```python
plt.hist(a,color = "#001256")
```

```python
plt.hist(a,color = "#001256",bins = 2)
```

```python
scores = [68, 75, 84, 92, 78, 88, 76, 85, 90, 87, 72, 80, 95, 81, 79, 73, 89, 82, 71, 77]

# Define the number of bins and their range
num_bins = 5  # Number of bins
bin_width = (max(scores) - min(scores)) / num_bins  # Calculate bin width
bins = [min(scores) + i * bin_width for i in range(num_bins + 1)]  # Define bin edges
print(bins)
# Create the histogram
plt.hist(scores, bins=bins, edgecolor='k', alpha=0.7)

# Add labels and title
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('Histogram of Exam Scores')

# Show the histogram
plt.show()
```

# BOXPLOT

```python
class_A_scores = [75, 80, 85, 90, 95, 100]
class_B_scores = [60, 65, 70, 75, 80, 85, 90,110,120]

# Combine the data into a list
data = [class_A_scores, class_B_scores]

# Create a boxplot
plt.boxplot(data, labels=['Class A', 'Class B'])

# Add labels and title
plt.xlabel('Classes')
plt.ylabel('Exam Scores')
plt.title('Boxplot of Exam Scores for Two Classes')

# Show the boxplot
plt.show()
```

# VIOLINPLOT

```python
class_A_scores = [75, 80, 85, 90, 95, 100]
class_B_scores = [60, 65, 70, 75, 80, 85, 90,110,120]

# Combine the data into a list
data = [class_A_scores, class_B_scores]

# Create a boxplot
plt.violinplot(data)

# Add labels and title
plt.xlabel('Classes')
plt.ylabel('Exam Scores')
plt.title('Boxplot of Exam Scores for Two Classes')

# Show the boxplot
plt.show()
```

# PIECHART

A pie chart is a circular graph that represents data as slices or wedges of a pie. Each slice represents a portion or percentage of the whole, and the entire pie represents 100%. Pie charts are commonly used to show the composition or distribution of a categorical dataset. 


```python
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 35, 20, 10]

plt.figure(figsize=(6, 6))  
plt.pie(values, labels=categories)
#plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title('Sample Pie Chart')

plt.show()
```

```python
#Highligthing the slice

plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140,explode = (0.5,0.3,0.2,0.1))
#plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140,explode = (0.5,0.3,0.2,0.1),colors = ["grey","blue","orange","pink","cayon"])
```

```python
plt.pie(values, labels=categories, autopct='%1.1f%%', radius = 3)
plt.pie([5],colors = 'w',radius = 2)
plt.show()
```
